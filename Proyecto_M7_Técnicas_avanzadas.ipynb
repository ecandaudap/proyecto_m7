## Actividad 2. Entrenamiento del modelo

### Función de limpieza de datos, tokenización y eliminación de palabras de parada.

```python
stop_words_english = set(stopwords.words('english')) # se selecciona el idioma
stop_words_english.update(["game", "im", "ive", "app", "phone", "also", "like",
                           "get", "would", "even", "play", "much", "many",
                           "please", "fix", "full", "review", "every",
                           "angry", "birds", "candy", "crush"])

def cleantext(text):

    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    text_wo_url = url_pattern.sub('', text) # Remover URLs
    text_wo_numbers = re.sub(r'\d+', '', text_wo_url) # Remover números
    text_english_only = re.sub(r'[^a-zA-Z0-9\s.,;\'"-]', '', text_wo_numbers) # Mantener solo letras inglesas y quitar demás caractéres
    text_wo_punct = text_english_only.translate(str.maketrans('', '', string.punctuation)) # Quitar signos de puntuación
    text_lower = text_wo_punct.lower() # Convertir a minúsculas
    text_wo_blank = text_lower.strip() # Eliminar espacios en blanco
    text_tokenized = nltk.word_tokenize(text_wo_blank) # Tokenizar
    tokens =[]
    for token in text_tokenized:
        if token in stop_words_english:
            continue
        tokens.append(token)
    return tokens


# Limpieza de texto
df_games['reviews_without_stopwords'] = df_games['Translated_Review'].apply(cleantext)

df_games
```

### Generar Nube de palabras de acuerdo al Sentimiento del Comentario

```python
for category in df_games['Sentiment'].unique():
    dfg=[]
    for token_list in df_games[df_games['Sentiment'] == category]['reviews_without_stopwords']:
        dfg += token_list
    word_freq = Counter(list(bigrams(dfg)))
    bigram_dict = {}
    for n in word_freq:
        key = ' '.join(n)
        value = word_freq[n]
        bigram_dict[key] = value
    wordcloud = WordCloud(width=800, height=400, background_color='white', colormap = 'Paired', collocations=True, max_words=30).generate_from_frequencies(bigram_dict)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(category)
    plt.axis('off')
    plt.show()
```

### Los Bigramas más comunes por sentimiento

```python
# Iterar sobre categorías únicas de sentimiento
for category in df_games['Sentiment'].unique():
    dfg = []

    # Recopilar tokens para la categoría actual
    for token_list in df_games[df_games['Sentiment'] == category]['reviews_without_stopwords']:
        dfg += token_list

    # Calcular frecuencia de bigramas
    word_freq = Counter(list(bigrams(dfg)))

    # Crear diccionario de bigramas
    bigram_dict = {}
    for n in word_freq:
        key = ' '.join(n)
        value = word_freq[n]
        bigram_dict[key] = value

    # Graficar los 10 bigramas más comunes
    top_bigrams = dict(sorted(bigram_dict.items(), key=lambda item: item[1], reverse=True)[:10])

    plt.figure(figsize=(10, 6))
    plt.bar(top_bigrams.keys(), top_bigrams.values())
    plt.title(f'Bigramas más comunes por Sentimiento: {category}')
    plt.xlabel('Bigramas')
    plt.ylabel('Frequencia')
    plt.xticks(rotation=90)
    plt.show()
```

El resultado del análisis muestra que los sentimientos negativos sobre la categoría "GAMES" se centran en quejas sobre la pérdida de tiempo o el tiempo de espera, la publicidad, el dinero y el bajo rendimiento del juego.

Los sentimientos positivos, en cambio, consisten en muchos elogios (diversión, bueno, divertido, genial para pasar el tiempo), así como el buen desempeño del mismo.


### PCA y Similitud de Coseno entre vectores de palabras.

Tal y como se hizo en el curso, el modelo de vectorización recibirá el nombre de `model_mini` y para estimarlos se utilizará la clase `Word2Vec`, cuyos parámetros se explican el anexo del texto.

Para este ejercicio se utilizará un vec_size de 100, con la intención de capturar la mayor cantidad posible de información en los vectores a obtener (1 x 100), sin embargo los valores típicos pueden oscilar entre 50 y 300, dependiendo del caso. Asimismo, se mantuvo en  1 el conteo mínimo para que cada palabra tenga su propio vector.

```python
def plot_pca_explained_variance(df, sentiment):
    data = df[df['Sentiment'] == sentiment]['reviews_without_stopwords']

    model_mini = Word2Vec(sentences=data, min_count=1, vector_size=100)
    X_mini = model_mini.wv[model_mini.wv.index_to_key]

    pca = PCA(n_components=2)
    X_mini_2d = pca.fit_transform(X_mini)

    explained_variance_ratio = pca.explained_variance_ratio_
    cumulative_explained_variance = np.cumsum(explained_variance_ratio)

    df_cumulative_explained_variance = pd.DataFrame({
        'Componente Principal': np.arange(1, len(cumulative_explained_variance) + 1),
        'Proporción Acumulada de Varianza Explicada': cumulative_explained_variance
    })

    plt.figure(figsize=(8, 6))
    plt.bar(df_cumulative_explained_variance['Componente Principal'], df_cumulative_explained_variance['Proporción Acumulada de Varianza Explicada'])
    plt.suptitle('Análisis de Componentes Principales (PCA)', fontsize=14)
    plt.title(f'({sentiment.capitalize()} Sentiment)')
    plt.xlabel('Número de Componentes Principales')
    plt.ylabel('Proporción Acumulada de Varianza Explicada')
    plt.grid(True)
    plt.show()

# Obtener categorías únicas de sentimiento
sentiments = df_games['Sentiment'].unique()

# Iterar sobre cada sentimiento y generar el gráfico
for sentiment in sentiments:
    plot_pca_explained_variance(df_games, sentiment)
```

De los resultados anteriores se deduce que, mientras que la explicación de comportamiento de los sentimientos positivos y negativos se pueden capturar con tan solo 2 variables esto no se logra en el caso de los neutros.

Esto se debe a que es mucho más difícil capturar el sentido de un mensaje considerado neutro, ya que se necesitan muchos más elementos que cuando se considera positivo o negativo.

```python
def plot_word_similarity(sentiment, df, words_to_compare):
    # Filtrar el DataFrame por el sentimiento específico
    filtered_df = df[df['Sentiment'] == sentiment]

    # Crear el modelo Word2Vec
    model = Word2Vec(sentences=filtered_df['reviews_without_stopwords'], min_count=1, vector_size=100)

    # Obtener vectores de palabras
    vectors = [model.wv[word] for word in words_to_compare]

    # Calcular la similitud coseno
    similarity_matrix = cosine_similarity(vectors, vectors)

    # Crear un mapa de calor para visualizar la similitud
    plt.figure(figsize=(10, 8))
    plt.imshow(similarity_matrix, cmap='viridis', interpolation='nearest')
    plt.colorbar()
    plt.xticks(range(len(words_to_compare)), words_to_compare, rotation=45, ha='right')
    plt.yticks(range(len(words_to_compare)), words_to_compare)
    plt.title(f'Similitud Coseno entre Vectores de Palabras ({sentiment.capitalize()} Sentiment)')
    plt.show()

# Obtener categorías únicas de sentimiento
sentiments = ['Negative', 'Positive']

# Palabras para comparar la similitud
words_to_compare = ['time', 'money', 'spend', 'fun', 'annoying', 'ads', 'level', 'free', 'love']

# Iterar sobre cada sentimiento y generar el gráfico de similitud
for sentiment in sentiments:
    plot_word_similarity(sentiment, df_games, words_to_compare)
```

En la celda anterior se relacionan los sentimientos del comentario con una lista de palabras para comparar la similitud del coseno. El resultado muestra un mapa de calor para visualizar la similitud entre las palabras especificadas.

La similitud del coseno es una medida numérica que cuantifica la similitud entre dos vectores en un espacio vectorial. En el contexto de procesamiento de lenguaje natural y modelos de incorporación de palabras como Word2Vec, la similitud del coseno se utiliza comúnmente para medir la proximidad semántica entre palabras.

La similitud del coseno produce un valor en el rango [-1, 1]. Un valor de 1 indica similitud perfecta, -1 indica máxima disimilitud, y 0 indica independencia lineal (es decir, los vectores son ortogonales).

Las matrices muestran en el caso de los sentimientos negativos, las palabras ***time*** y ***money***, se relacionan fuertemente con ***annoying***, ***ads*** y ***level***.
***spend*** y ***fun** tiene, en general las relaciones positivas más bajas e incluso, podrían tener una relación negativa con **free**.
